## Part 1: The Discovery

### A Holiday Booking Gone Wrong

**üìñ Read the full story:** [Preface](invisible-users/preface.md)

In late 2024, author Tom Cranstoun tried something simple: he asked an AI assistant to compare tour operators for a Southeast Asia trip.

The AI confidently reported: "Tour Company A offers a 14-day experience covering Vietnam, Cambodia, and Thailand. Tour Company B only covers one day in Bangkok. Company A is clearly superior."

**This was completely wrong.**

Company B also had a 14-day itinerary. But they split it across 14 pages, each requiring a click on "Next" to view the next day. The AI agent saw Day 1, assumed that was everything, and gave useless advice.

Tom ended up booking with a different company‚Äîone that showed their entire itinerary on a single scrollable page. The agent could read it. Tom could compare it. That company got his money.

**The paginated site lost a customer and will never know why.**

## Part 2: Understanding Invisible Failures

**üìñ Deep dive:** [Chapter 2: The Invisible Failure](invisible-users/chapter-02-the-invisible-failure.md)

### What Makes Users "Invisible"?

AI agents are called "invisible users" for two reasons:

1. **They're invisible to site owners** ‚Äî Unless you're specifically tracking agent traffic (and most aren't), you have no idea how many agents visit your site or whether they succeed. They blend into analytics as slightly unusual sessions.

2. **Your interface is partly invisible to them** ‚Äî They can't see your beautiful animations, don't notice subtle color changes, miss three-second toast notifications, and don't understand that a loading spinner means "wait."

### The Five Types of Invisible Failure

Let's explore the specific patterns that break AI agents:

### Understanding Agent Diversity

When I refer to "AI agents" throughout this notebook and book, I'm describing a diverse ecosystem with varying capabilities:

**Server-based agents** (ChatGPT, Claude via API) access websites remotely from cloud infrastructure, without any inherited browser state or authentication.

**CLI agents** (Claude Code, Cline) run as command-line tools on your local machine but fetch web content remotely - no session inheritance.

**Browser agents** (Playwright, Selenium) use full browser automation to execute JavaScript and handle dynamic content, but they're slower and resource-intensive.

**Browser extension assistants** (ChatGPT sidebar, Claude browser extension) run inside your browser and inherit your authenticated sessions, cookies, and proof-of-humanity tokens.

**Local (SMOL) agents** are lightweight tools running entirely on your device with privacy-focused approaches.

**IDE-integrated browser controls** (Google Antigravity) combine development environment features with browser capabilities.

**Why this matters:** The failures described in this book affect these agents differently:

- **Session access** varies - browser extensions inherit your authenticated sessions whilst external agents must authenticate independently
- **JavaScript execution** differs - some agents run full browsers whilst others parse only static HTML
- **State detection** capabilities range from sophisticated to simple "URL didn't change = nothing happened"
- **Authentication** approaches differ fundamentally

**The critical insight:** We cannot design for just one agent type. A pattern that works brilliantly for browser extensions might fail completely for server-based agents. This book focuses on universal patterns that work across the entire ecosystem.

Throughout the book, when agent type matters, I'll specify which type. When I refer to "agents" generally, I mean patterns affecting the entire ecosystem.

## Part 3: The Accessibility Connection

**üìñ Explore further:** [Chapter 1: What You Will Learn](invisible-users/chapter-01-what-you-will-learn.md)

### The Curb Cut Effect

The web accessibility movement taught us something important: **designing for edge cases improves experience for everyone.**

- **Curb cuts** help wheelchair users, parents with pushchairs, travelers with suitcases, and delivery workers
- **Captions** help deaf viewers, people in noisy environments, language learners, and anyone who prefers reading
- **High-contrast text** helps visually impaired users and people reading screens in bright sunlight

Agent-friendly design follows the same principle.

When you make your checkout flow work for an AI agent, you're also making it work for:
- The stressed parent trying to complete a purchase while distracted
- The elderly user who misses rapid notifications
- The person with ADHD who can't track state changes across paginated screens
- The screen reader user who can't navigate visual hierarchy

### Why Accessibility Standards Already Cover This

**WCAG (Web Content Accessibility Guidelines)** already addresses many agent-breaking patterns:

**WCAG 2.2 Success Criterion 2.2.1 (Timing Adjustable, Level A):**
> "For each time limit that is set by the content, users can turn off, adjust, or extend the time limit."

This covers disappearing toast notifications‚Äîthey're timing-based content that vanishes before users can process it.

**WCAG 2.2 Success Criterion 3.3.1 (Error Identification, Level A):**
> "If an input error is automatically detected, the item that is in error is identified and the error is described to the user in text."

This requires persistent, text-based error messages‚Äînot visual-only indicators.

**WCAG 2.2 Success Criterion 1.3.1 (Info and Relationships, Level A):**
> "Information, structure, and relationships conveyed through presentation can be programmatically determined."

This means semantic HTML, proper ARIA attributes, and structured data‚Äîall things that help AI agents parse content.

**The problem:** Many sites treat accessibility as a compliance checkbox rather than understanding why these guidelines exist.

**The opportunity:** Commercial pressure from agent failures might finally drive meaningful accessibility improvements.

## Part 4: The Business Reality

**üìñ Full analysis:** [Chapter 4: The Business Reality](invisible-users/chapter-04-the-business-reality.md)

### Why This Matters Now

AI agents aren't a future concern. They're here.

People already use ChatGPT, Claude, and other AI assistants to:
- Research products and services
- Compare options and prices
- Check availability
- Complete forms
- Some are authorizing agents to make bookings and purchases

**And most of it is failing.**

Not dramatically. Not with error messages. Failing **quietly**:
- Agents report success when they've actually failed
- Agents miss information hidden behind pagination
- Agents make recommendations based on incomplete data
- Agents give up on tasks humans could complete easily

### The Invisible Impact

Your analytics won't show this. Your error logs won't capture it. Your customer support won't hear about it.

**The agent fails, the human moves on to a competitor's site, and you never know the sale was lost.**

**Traditional conversion optimization assumes:**
1. User arrives on your site
2. User navigates your interface
3. User completes or abandons the task
4. You see the funnel drop-off and optimize

**Agent-mediated reality:**
1. Agent arrives on your site
2. Agent attempts to parse your content
3. Agent silently fails (e.g., misses paginated content)
4. Human never visits‚Äîsees agent's report and goes elsewhere
5. **You never see them in your analytics**

This is fundamentally different from traditional optimization. You're not optimizing a funnel you can measure‚Äîyou're fixing invisible problems that cost you customers you never knew existed.

## Part 5: The Content Creator's Dilemma

**üìñ Complete analysis:** [Chapter 5: The Content Creator's Dilemma](invisible-users/chapter-05-the-content-creators-dilemma.md)

### When Agents Extract Without Visiting

AI agents don't just struggle with websites‚Äîthey're changing the economics of content creation.

**The traditional model:**
1. Publisher creates content
2. User visits website
3. User views ads or subscribes
4. Publisher gets revenue

**The agent-mediated model:**
1. Publisher creates content
2. Agent reads content
3. Agent summarizes for user
4. User never visits site
5. Publisher gets nothing

This isn't theoretical. Google's AI Overviews, ChatGPT's browsing, and Claude's web research all extract information without sending traffic.

### The Attention Economy Under Pressure

Ad-funded content depends on:
- Page views
- Time on site
- Ad impressions

When agents consume content without visiting, all three metrics collapse to zero.

**Implications:**
- News sites lose revenue
- Recipe blogs can't monetize
- Tutorial content becomes free for everyone except the creator
- Only subscription models survive

**Example:** A recipe blogger spends hours creating unique recipes, photography, and instructions. Google's AI Overview extracts the recipe, displays it directly in search results. User never visits the blog. No ad revenue. No affiliate link clicks. The creator's work becomes free content for Google to monetize through their own ads.

## Part 6: The Security Maze

**üìñ Technical deep dive:** [Chapter 6: The Security Maze](invisible-users/chapter-06-the-security-maze.md)

### Authentication and Session Inheritance

When you ask an AI agent to check your bank balance, something subtle happens: **the agent inherits your authenticated session.**

You're logged into your bank. The agent opens the page using your browser context. As far as the bank knows, it's you.

### The Verification Paradox

Banks can't distinguish between:
- You accessing your account
- An AI agent accessing your account on your behalf
- Malware accessing your account

All three look identical from the server's perspective.

**Current security measures fail:**
- **Two-factor authentication**: You've already authenticated; the agent inherits the session
- **Behavioral analysis**: Agent behavior looks suspicious but you authorized it
- **Device fingerprinting**: Same device, same browser, same session
- **Geographic checks**: Same location as legitimate user

### Real-World Scenario

**User:** "Agent, please check my bank balance and tell me my recent transactions."

**What happens:**
1. Agent opens banking site in user's browser context
2. Session cookies authenticate automatically
3. Agent reads balance: ¬£5,420.32
4. Agent reads transactions: recent purchases, bills
5. Bank has no idea this was an agent, not the user

**Security implications:**
- No audit trail distinguishing human from agent actions
- No way to revoke agent access without logging out entirely
- No granular permissions ("read balance" vs "transfer funds")
- If agent is compromised, full account access is compromised

**The fundamental problem:** Financial security assumes direct human‚Üísite interaction. Agent-mediated access breaks that assumption entirely.

## Part 9: Designing for Both

**üìñ Solutions and patterns:** [Chapter 9: Designing for Both](invisible-users/chapter-09-designing-for-both.md)

### The Convergence Principle

Good design converges. When you design for one edge case well, you often improve the experience for many others.

**The Convergence Principle states:**

The patterns that work for AI agents are the same patterns that work for:
- Screen reader users
- Keyboard-only navigation
- People with cognitive disabilities
- Users under stress or distraction
- Anyone in non-ideal conditions

### Core Design Principles

1. **Clear State, Always Visible**
   - Make loading, success, and error states explicit
   - Don't rely on color or animation alone
   - Use data attributes and ARIA labels

2. **Persistent Feedback**
   - Error messages that stay visible
   - Success confirmations that don't disappear
   - Clear indication of what happened

3. **Complete Information**
   - Show full pricing upfront
   - Avoid "from ¬£99" with hidden costs
   - Make all content accessible without pagination

4. **Semantic Structure**
   - Use proper HTML elements
   - Add JSON-LD structured data
   - Clear heading hierarchy

5. **URL-Based State**
   - URL should reflect current state
   - Bookmarkable, shareable
   - Browser history works correctly

## A Critical Responsibility

This book addresses **two audiences**: developers who build websites, and people who use AI agents to interact with them. Both have distinct responsibilities.

### For Developers: Fix What You Can Control

The technical patterns throughout this book‚Äîclear state, persistent errors, semantic structure, transparent pricing‚Äîimprove experiences for everyone. When you implement agent-friendly design, you're not just helping AI systems. You're helping:

- Screen reader users who rely on semantic HTML
- Elderly users who miss disappearing notifications  
- People with cognitive disabilities who struggle with hidden state
- Users under stress who need explicit feedback

**Your responsibility:** Implement patterns that work for all users. Use the Priority 1 quick wins from Chapter 9. Test with accessibility tools. Fix the invisible failures.

### For Users: Careful Prompts for High-Stakes Services

**Not everything can be fixed with better design.**

Throughout this book, you've encountered three categories that don't have technical solutions you can implement with HTML:

1. **Banking security** (Chapter 6) ‚Äî Session inheritance, verification paradoxes, unclear liability
2. **Legal liability** (Chapter 7) ‚Äî Grey zones, undefined frameworks, cross-border complexity  
3. **Speculative commerce** (Chapter 4) ‚Äî Dark warehouses, identity delegation, business model disruption

These aren't problems developers fix with code. They're environments where **the instructions you give to AI agents matter enormously**.

### The High-Stakes Scenarios

**Banking:** When you authorize an agent to access your bank account, you're accepting risk. If the agent misinterprets "transfer ¬£500 to my savings" as "transfer ¬£5,000," you may be liable. The bank's security systems can't distinguish you from your agent.

**Legal services:** When an agent fills out legal forms or interprets contracts on your behalf, mistakes have consequences. Ambiguous instructions like "agree to standard terms" could commit you to obligations you didn't intend.

**Commercial transactions:** When agents make purchases, book services, or enter contracts, you're responsible for the outcomes. "Book the cheapest flight" might mean the flight with the worst cancellation policy.

### What This Means for Agent Users

**Be specific with prompts.** Don't say "pay my bills"‚Äîspecify which bills, how much, when. Don't say "book a hotel"‚Äîprovide dates, location, price range, requirements.

**Verify before authorizing.** Review what the agent plans to do before it executes. Check confirmations independently. Don't assume success just because the agent reports success.

**Understand your liability.** When you delegate authority to an agent, you're accepting responsibility for its actions. Agent platforms typically disclaim liability for mistakes. You're on your own if something goes wrong.

**Use appropriate authorization levels.** Grant minimal access needed. Read-only for information gathering. Limited transaction authority for specific purchases. Never share credentials unless absolutely necessary.

### The Dual Responsibility

Agent-friendly design helps. Clear interfaces reduce ambiguity. Persistent feedback prevents silent failures. Transparent pricing avoids surprises.

But careful, considered use of agents is equally important. 

**Developers:** Build clarity into your interfaces. Fix the patterns that break for everyone.

**Users:** Craft precise prompts when stakes are high. Verify outcomes independently. Understand the liability you're accepting.

This book addresses both. Technical patterns make systems clearer. User awareness prevents costly mistakes.

**The future of agent-mediated interaction depends on both responsibilities being taken seriously.**
